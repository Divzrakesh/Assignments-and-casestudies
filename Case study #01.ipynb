# 1. Read the dataset to the python environment


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cars_data = pd.read_csv('cars_data.csv')
cars_data
```

# 2. Check for the null values present in the dataset


```python
print(cars_data.isna().sum())
```

# 3. Bar graph of male vs female buyers participated in the sales


```python
tot =cars_data['Buyer Gender'].value_counts()
tot
```

```python
c = ['red','blue']
cars_data['Buyer Gender'].value_counts().plot(kind='bar',color = c);
plt.title('Male Vs Female Buyers')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.grid(linestyle='--')
plt.show()
```

```python
df = cars_data[['Make','Model','Sale Price']].sort_values(by=['Sale Price'],ascending=False)
df.head(5)
```
